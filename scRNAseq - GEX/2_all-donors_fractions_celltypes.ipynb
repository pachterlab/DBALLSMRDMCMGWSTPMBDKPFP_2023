# Show fraction of batch/donor in each Leiden cluster

```python
def plot_fraction_per_sample(
    adata,
    groups="leiden",
    fractions="title",
    colors=None,
    frac_order=None, 
    group_order=None,
    legend_cols=1,
    normalize_to_total=False,
    title=None,
    xticklabel_rotation=0,
    figsize=None,
    save=False
):
    ## Save fractions to plot in dataframe
    # # Convert sample column to string (instead of categorical)
    # adata.obs[fractions] = adata.obs[fractions].astype(str)

    df_normalized = pd.DataFrame()

    if normalize_to_total:
        # Normalize each celltype count to total number of cells in that batch by dividing
        df_normalized = ((adata.obs.groupby(groups)[fractions].value_counts()/ adata.obs.groupby(fractions)["species"].count())
            .unstack()
            .fillna(0)
        )
        # Convert columns to string
        df_normalized.columns = df_normalized.columns.astype(str)
        # Get total normalized count
        df_normalized["total_normalized_count"] = df_normalized.sum(axis=1).values

        # Compute fraction of normalized total count for each sample
        for sample in adata.obs[fractions].unique():
            df_normalized[f"{sample}_fraction"] = (df_normalized[sample] / df_normalized["total_normalized_count"]).values

    else:
        # Do not normalize to total count (use if we do not expect these to be equally distributed across the clusters)
        df_normalized = (adata.obs.groupby(groups)[fractions].value_counts().unstack().fillna(0))
        df_normalized["total_count"] = df_normalized.sum(axis=1).values

        for sample in df_normalized.columns[:-1]:
            df_normalized[f"{sample}_fraction"] = (df_normalized[sample] / df_normalized["total_count"]).values

    # Get total cellcount
    df_normalized["total_cellcount"] = (adata.obs.groupby(groups).size().values.astype(int))
    
    # Sort data
    if group_order:
        df_normalized = df_normalized.reindex(group_order)
    else:
        df_normalized = df_normalized.sort_values(list(df_normalized.filter(like='_fraction').columns.values))

    ## Plot
    if figsize:
        fig, ax = plt.subplots(figsize=figsize)
    else:
        fig, ax = plt.subplots(figsize=(15, 5))
    width = 0.75
    alpha = 1
    fontsize = 14

    clusters = df_normalized.index.values
    cellcounts = df_normalized["total_cellcount"].values
    
    previous_samples = 0
    if frac_order:
        for c_idx, sample in enumerate(frac_order):
            ax.bar(
                clusters,
                df_normalized[f"{sample}_fraction"].values,
                width,
                bottom=previous_samples,
                color=colors[c_idx] if colors else None,
                label=sample,
                alpha=alpha,
            )
            previous_samples = previous_samples + df_normalized[f"{sample}_fraction"].values
    else:
        for c_idx, sample in enumerate(adata.obs[fractions].unique()):
            ax.bar(
                clusters,
                df_normalized[f"{sample}_fraction"].values,
                width,
                bottom=previous_samples,
                color=colors[c_idx] if colors else None,
                label=sample,
                alpha=alpha,
            )
            previous_samples = previous_samples + df_normalized[f"{sample}_fraction"].values

    # Add value above each bar
    for index, value in enumerate(cellcounts):
        ax.text(x=index, y=1.01, s=int(value), size=fontsize - 3, ha="center")

    if xticklabel_rotation == 0:
        ax.set_xticklabels(clusters)
    else:
        ax.set_xticklabels(clusters, rotation=xticklabel_rotation, ha="right")

    ax.legend(bbox_to_anchor=(1.001, 1.025), loc="upper left", fontsize=fontsize, ncols=legend_cols)

    ax.set_title(title, y=1.0, pad=20, fontsize=fontsize + 2)
    ax.set_ylabel("Fraction of cells", fontsize=fontsize)
    if "leiden" in groups:
        ax.set_xlabel("Leiden cluster", fontsize=fontsize)
    else:
        ax.set_xlabel(groups.capitalize(), fontsize=fontsize)
    ax.tick_params(axis="both", labelsize=fontsize)

    # ax.axhline(y=0.25, color="grey", linestyle="--")
    # ax.axhline(y=0.5, color="grey", linestyle="--")
    # ax.axhline(y=0.75, color="grey", linestyle="--")

    ax.margins(x=0.01, y=0.06)
    ax.grid(False)

    plt.tight_layout()
    
    if save and normalize_to_total:
        plt.savefig(
            f"figures/2_{groups}_{fractions}_norm_fractions.png",
            dpi=300,
            bbox_inches="tight",
        )
        plt.savefig(
            f"figures/2_{groups}_{fractions}_norm_fractions.svg",
            dpi=300,
            bbox_inches="tight",
        )
    elif save:
        plt.savefig(
            f"figures/2_{groups}_{fractions}_fractions.png",
            dpi=300,
            bbox_inches="tight",
        )
        plt.savefig(
            f"figures/2_{groups}_{fractions}_fractions.svg",
            dpi=300,
            bbox_inches="tight",
        )

    fig.show()
```

The “2_leiden_timepoint_fractions” figure is interesting because it shows how the timepoints are separated by Leiden clustering, meaning that the gene expression is different between the timepoints (this would be the equivalent of showing a UMAP where the timepoints separate into different clusters, but you can actually read the fractions quantitatively instead of having to guess the number of dots). IP clusters out entirely (not surprisingly), but you can also see that day 14 and 28 are quite different from day 0, while day 90 often clusters together with day 0. This is based on gene expression across all donors and cell types. There is also an indication of data quality here with days that should be similar clustering together, e.g. days -12 and -6.

2_leiden_sample_fractions” shows the breakdown of the batches/datasets per Leiden cluster and shows much less of a clear separation compared to the timepoints. This supports the decision not to do batch correction.

```python
marker_df = pd.read_csv("../COH/marker_lists/marker_list.csv")
# Only keep markers marked for cell type assignment
marker_df = marker_df[marker_df["celltype_assignment"]==True]
```

Get Ensembl IDs from marker gene symbols using gget:

```python
# Get Ensembl IDs from gene symbols using gget
search_results = gget.search(marker_df["gene_symbol"].values, species="human")
# Filter gget results for exact matches and protein-coding biotypes
search_results_filtered = search_results[search_results["gene_name"].isin(marker_df["gene_symbol"].values)][search_results["biotype"]=="protein_coding"]
```

Why is there more than one Ensembl ID per gene symbol?  
"When you’re building a genome assembly for a single species with very diverse populations or strains, it is difficult to capture all of that lovely variability in a single reference sequence. That’s where haplotypic regions come in.
Haplotyptic regions are defined by the Genome Reference Consortium (GRC) and are visible on all their genome assemblies for human, mouse, and zebrafish. They can also be called ‘alternate locus’, ‘partial chromosomes’, and ‘alternate alleles’ or ‘assembly exceptions’ (although this definition also includes patches). Most often these are added because through sequencing a variety of individuals different versions of regions are identified. These regions can have small sequence differences, contain different gene structures or different genes entirely, or contain genes in a different order compared to the reference genome assembly.
We display haplotypic regions in Ensembl as red stripes overlapping the chromosome, and you can easily compare these with the reference assembly. The haplotypes themselves have a unique identifier, which links them to the region on the primary assembly. In human these start CHR_HSCHR followed by the chromosome number. (https://www.facebook.com/151892474871670/posts/tipoftheweek-seeing-red-in-the-genome-haplotypes-and-haplotypic-regions-when-you/2177783548949209/)"
https://www.researchgate.net/post/How-to-deal-with-multiple-ensemble-IDs-mapping-to-one-gene-symbol-in-a-RNA-Seq-dataset  

See the different chromosome regions for the Ensembl IDs for gene symbol NCR1 as an example:

```python
# Add Ensembl IDs to marker gene df
marker_df["ensembl_id"] = np.NaN

ensembl_id_list = []

for i, gene_symbol in enumerate(marker_df["gene_symbol"].values):
    ensembl_ids = search_results_filtered[
        search_results_filtered["gene_name"] == gene_symbol
    ]["ensembl_id"].values
    
    ensembl_id_list.append(ensembl_ids)
    
    if len(ensembl_ids) < 1:
        print(f"An Ensembl ID for gene {gene_symbol} could not be found.")

marker_df["ensembl_id"] = ensembl_id_list
marker_df
```

# Plot marker gene expression across Leiden clusters

```python
def flatten(xss):
    """
    Function to flatten array of arrays.
    """
    return [x for xs in xss for x in xs]
```

Add gene names of marker genes to adata:

```python
adata.var["gene_symbol"] = [np.NaN] * len(adata.var)

for ensembl_ids, gene_symbol in zip(marker_df["ensembl_id"].values, marker_df["gene_symbol"].values):
    if len(ensembl_ids)>1:
        for ensembl_id in ensembl_ids:
            adata.var.loc[adata.var.index == ensembl_id, "gene_symbol"] = gene_symbol
    else:
        adata.var.loc[adata.var.index == ensembl_ids[0], "gene_symbol"] = gene_symbol
```

Define function to create a short marker df containing only the markers that are actually contained in adata:

```python
celltype_order = ["T cells", "B cells", "Natural killer cells", "Mast cells", "Dendritic cells", "Plasmacytoid dendritic cells", "Monocytes"]
celltype_labels = ["T cells", "B cells", "Natural killer cells", "Mast cells", "Dendritic cells", "Plasmacytoid\ndendritic cells", "Monocytes"]
```

```python
def find_markers(adata):  
    # Create a copy of the original data
    adata_test = adata.copy()

    # Find indices of all marker genes
    ens_idx = np.isin(
        adata_test.var.index.values, flatten(marker_df["ensembl_id"].values)
    )

    # Slice adata_test based on these indices
    adata_test = adata_test[:, ens_idx].copy()
    
    # Create data frame keeping only markers/IDs that are also present in the data
    marker_df_short = pd.DataFrame()
    marker_df_short["ensembl_id"] = adata_test.var.index.values
    marker_df_short["gene_symbol"] = adata_test.var.gene_symbol.values

    for gene_symbol in marker_df_short["gene_symbol"].values:
        celltype = marker_df[marker_df["gene_symbol"] == gene_symbol]["celltype"].values[0]
        marker_df_short.loc[marker_df_short.gene_symbol == gene_symbol, "celltype"] = celltype

    # Sort marker by celltype and gene symbol
    marker_df_short["celltype"] = pd.Categorical(marker_df_short.celltype, categories=celltype_order, ordered=True)
    marker_df_short = marker_df_short.sort_values(["celltype", "gene_symbol"]).reset_index(drop=True)

    return marker_df_short
```

Define function to plot marker gene expression per group:

```python
def plot_marker_gene_exp(adata, var_grouping, var_labels, var_positions, groupby, dendrogram=True, save=False):
    marker_df_short = find_markers(adata)

    markers = ["CAR"] + list(marker_df_short["ensembl_id"].values)
    temp_var_labels = ["", "", "", "", "", "", ""]

    ax = sc.pl.dotplot(
        adata,
        markers,
        groupby=groupby,
        use_raw=True,
        dendrogram=dendrogram,
        var_group_positions=var_grouping,
        var_group_labels=temp_var_labels,
        var_group_rotation=45,
        swap_axes=False,
        standard_scale=None,
        show=False,
    )
    
    # ax["mainplot_ax"].set_title(f"Expression of marker genes in {'-'.join(adata.obs['tissue'][0].split('/'))} dataset", pad=120)
    
    ## Rotate var labels
    # Get initial label positions from: ax["gene_group_ax"].properties()["children"]
    for position, label in zip(var_positions, var_labels):
        ax["gene_group_ax"].text(position, 1.1, label, rotation=45, ha="left")

    ax["mainplot_ax"].set_xticklabels(
        ["CAR"] + list(marker_df_short["ensembl_id"].values + " - " + marker_df_short["gene_symbol"].values.astype(str)), 
        rotation=45, ha="right",)
    ax["mainplot_ax"].set_xlabel("Gene", fontsize=fontsize)
                                    
    if "leiden" in groupby:
        ax["mainplot_ax"].set_ylabel("Leiden cluster", fontsize=fontsize)
    else:
        ax["mainplot_ax"].set_ylabel(groupby.capitalize(), fontsize=fontsize)

    plt.tight_layout()
                                    
    # Save plot
    if save:
        plt.savefig(f"figures/2_{groupby}_marker-gene-expression.png", dpi=300, bbox_inches="tight")
        plt.savefig(f"figures/2_{groupby}_marker-gene-expression.svg", dpi=300, bbox_inches="tight")
```
