# Compare new alignment to human ref genome v108 + CAR to previous alignment (ref genome v94)

```python
%config InlineBackend.figure_format = 'retina'
%load_ext blackcellmagic

import json
import matplotlib.pyplot as plt
import numpy as np
import anndata
```

___

# Load data

```python
tissues = ["CART", "PB"]

datasets = [
    ["UPN375_376_388_393_PSCA_CART_X5SCR_SE6080_SA58165"],
    [
        "UPN375_388_393_PB_B_X5SCR_SE6452_SA71978",
        "UPN376_388_403_PB_C_X5SCR_SE6452_SA71979",
        "UPN376_388_406_PB_E_X5SCR_SE6452_SA71981",
        "UPN388_406_PB_I_X5SCR_SE6438_SA71770",
        "UPN388_394_PSCA_PB_M_X5SCR_SE8003_SA157962",
        "UPN375_388_PSCA_PB_N_X5SCR_SE8003_SA157963",
        "UPN375_388_PSCA_PB_O_X5SCR_SE8003_SA157964",
        "UPN388_394_PSCA_PB_P_X5SCR_SE8003_SA157965",
        "UPN388_394_PSCA_PB_Q_X5SCR_SE8003_SA157966",
        "UPN375_388_PSCA_PB_S_X5SCR_SE8003_SA157968",
    ],
]

samples = [["PSCA"], ["B", "C", "E", "I", "M", "N", "O", "P", "Q", "S"]]
```

Load kb run info:

```python
old_dict = {}

for idx, tissue in enumerate(tissues):
    for i, dataset in enumerate(datasets[idx]):
        f = open(f"../../../COH_data/kb/IRB17483_PSCA/{tissue}/{dataset}/run_info.json")
        old_dict[dataset] = json.load(f)
```

```python
new_dict = {}

for idx, tissue in enumerate(tissues):
    for i, dataset in enumerate(datasets[idx]):
        f = open(f"../../../COH_data/alignment_v2/{dataset}/run_info.json")
        new_dict[dataset] = json.load(f)
```

Load h5ad objects:

```python
adatas_raw = []

for idx, tissue in enumerate(tissues):
    for i, dataset in enumerate(datasets[idx]):
        adata = anndata.read(f"../../../COH_data/alignment_v2/{dataset}/counts_unfiltered/adata.h5ad")
        print(dataset)
        print(adata)

        adata.obs["species"] = ["homo_sapiens"] * adata.n_obs
        adata.obs["tissue"] = [tissue] * adata.n_obs
        adata.obs["library_id"] = ["_".join(dataset.split("_")[-2:])] * adata.n_obs
        adata.obs["sample"] = [samples[idx][i]] * adata.n_obs

        adatas_raw.append(adata)
```

___
# Plot alignment percentages

```python
def flatten(l):
    """
    Flatten a list of lists.
    """
    return [item for sublist in l for item in sublist]
```

```python
p_pseudoaligned_old = []
p_pseudoaligned_new = []
for idx, tissue in enumerate(tissues):
    for i, dataset in enumerate(datasets[idx]):
        p_pseudoaligned_old.append(old_dict[dataset]['p_pseudoaligned'])
        p_pseudoaligned_new.append(new_dict[dataset]['p_pseudoaligned'])
```

```python
labels = flatten(samples)
x = np.arange(len(labels))
width = 0.35
fontsize = 12

fig, ax = plt.subplots(figsize=(12,7))
rects1 = ax.bar(x - width / 2, p_pseudoaligned_old, width, label="94")
rects2 = ax.bar(x + width / 2, p_pseudoaligned_new, width, label="108 + CAR")

ax.set_ylabel("Percentage pseudoaligned", fontsize=fontsize)
# ax.set_title("Pseudoalignment", fontsize=fontsize+2)
ax.set_xticks(x, labels, fontsize=fontsize)
ax.legend(title="Ref genome version")

ax.bar_label(rects1, padding=2)
ax.bar_label(rects2, padding=2)

fig.tight_layout()

fig.savefig(f"figures/0_alginment_comparison.png", dpi=300, bbox_inches='tight')

fig.show()
```
