### Run VDJ alignment

```python
fastq_folders = [
    "../COH_data/alignment_v2/raw/VDJ/FT-SE6453",
    "../COH_data/alignment_v2/raw/VDJ/SE7663",
    "../COH_data/alignment_v2/raw/VDJ/SE8002",
    "../COH_data/alignment_v2/raw/VDJ/FT-SA58161"
]

outfolder = "../COH_data/alignment_v2/raw/VDJ_cellranger_results"
```

Note: Moved resulting folders to outfolder manually

```python
import glob
```

```python
%%time
for fastq_folder in fastq_folders:
    for folder in glob.glob(f"{fastq_folder}/FT-*/"):
        sample = "_".join(glob.glob(f"{folder}/*")[0].split("/")[-1].split("_")[:2])
        
        !$cellranger vdj \
            --localcores=60 \
            --id=$sample \
            --fastqs=$folder \
            --reference=../COH_data/reference_genome/cellranger/refdata-cellranger-vdj-GRCh38-alts-ensembl-5.0.0
```
